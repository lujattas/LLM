```python
from matplotlib import pyplot as plt
import numpy as np
from PIL import Image
import requests

urls = [
    "https://static.webteb.net/images/content/tbl_articles_article_25630_628a751a40d-43ae-41ad-851a-6ee1a3a010f5.jpg",
    "https://images.unsplash.com/photo-1530026454774-50cce722a1fb?w=800&auto=format&fit=crop&q=60&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8Mnx8Ym9keSUyMG9yZ2FufGVufDB8fDB8fHww",
    "https://plus.unsplash.com/premium_photo-1722525170821-75b4bec10224?w=800&auto=format&fit=crop&q=60&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8NXx8Ym9keSUyMG9yZ2FufGVufDB8fDB8fHww",
    "https://images.unsplash.com/photo-1615552440985-d652ebf4c199?w=800&auto=format&fit=crop&q=60&ixlib=rb-4.0.3&ixid=M3wxMjA3fDB8MHxzZWFyY2h8Mnx8ZXllJTIwb3JnYW58ZW58MHx8MHx8fDA%3D",
    "https://static.webteb.net/images/content/tbl_articles_article_21592_50092570062-36ec-4604-a9c0-3b60845388d6.jpg",
    "https://cdn.arageek.com/magazine/2021/04/encyclopedia-%D9%85%D8%A7-%D9%87%D9%88-%D8%AA%D8%AD%D9%84%D9%8A%D9%84-%D9%88%D8%B8%D8%A7%D8%A6%D9%81-%D8%A7%D9%84%D8%A8%D9%86%D9%83%D8%B1%D9%8A%D8%A7%D8%B3.jpg"






]

images = [
    Image.open(requests.get(url, stream=True).raw) for url in urls]

# let's see what we have
for image in images:
    plt.show(plt.imshow(np.asarray(image)))
```

```python
from transformers import CLIPProcessor, CLIPModel

model = CLIPModel.from_pretrained('openai/clip-vit-base-patch32')
processor = CLIPProcessor.from_pretrained('openai/clip-vit-base-patch32')
```

```python
captions = ["The heart is composed of four chambers; two atria and two ventricles",
            "Kidneys are paired bean-shaped organs placed retroperitoneally",
            "The lungs are the central organs of the respiratory system in humans",
            "The eyes are the organ responsible for vision in the body",
            "The liver is the largest solid internal organ in the human body, and the largest gland in the human body.",
            "The pancreas is a long, flat gland located horizontally behind the stomach"
          ]
inputs = processor(
    text=captions, images=images,
    return_tensors='pt', padding=True
)

outputs = model(**inputs)

probs = outputs.logits_per_image.argmax(dim=1)
probs
```

```python
for i, image in enumerate(images):
    argmax = probs[i].item()
    print(captions[argmax])
    plt.show(plt.imshow(np.asarray(image)))
```

```python


from transformers import pipeline
pipe = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
captions = []
for image in images:
  caption = pipe(image)[0]['generated_text']
  captions.append(caption)
print(captions)
```

```python
# prompt: Build a search engine for these images using a larger dataset of images

!pip install faiss-gpu

from sentence_transformers import SentenceTransformer, util
import torch

# Load a pre-trained SentenceTransformer model
embedder = SentenceTransformer('all-mpnet-base-v2')

# Embed the captions and images
image_embeddings = embedder.encode([np.array(image) for image in images], convert_to_tensor=True)
caption_embeddings = embedder.encode(captions, convert_to_tensor=True)

# Index the image embeddings for fast search
index = util.semantic_search(caption_embeddings, image_embeddings, top_k=1)

# Flatten the index
flat_index = [item[0]['corpus_id'] for item in index]

# Define a search function
def search(query):
  query_embedding = embedder.encode(query, convert_to_tensor=True)
  search_result = util.semantic_search(query_embedding, caption_embeddings, top_k=1)
  result_index = search_result[0][0]['corpus_id']
  return images[flat_index[result_index]]

# Example usage
query = "Show me the heart "
result_image = search(query)
plt.imshow(np.array(result_image))
plt.show()
```
